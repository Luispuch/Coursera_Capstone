```python
# Leer las tablas en pandas
wiki_tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')

# Imprimir el número total de tablas
print(f'Número total de tablas: {len(wiki_tables)}')

# Imprimir las primeras filas de todas las tablas en la página
for i in range(len(wiki_tables)):
    print('Tabla: ',i+1)
    print(wiki_tables[i].head())
```

```python
# Crear el dataframe con la tabla seleccionada
postal_code_df = wiki_tables[0]
print(postal_code_df.info())
postal_code_df.head()
```

```python
# definir las columnas del dataframe
column_names = ['Postal Code','Borough', 'Neighborhood'] 

# inicializar el dataframe
postal_code_canada = pd.DataFrame(columns=column_names)
postal_code_canada.head()
```

```python
#No correr 2 veces el código
#Crear el dataframe 
for i in range(9): 
   for j in range(20): 
        valor=postal_code_df.iloc[j,i]
        try:
            c=valor.split("(")[1].split(")")[0].replace(' /', ',')
        except:
            c='Not assigned'
        a=valor[slice(3)]
        b=valor[3:].split("(")[0]
        postal_code_canada.loc[len(postal_code_canada.index)] = [a, b, c]
postal_code_canada.head()
```

```python
# Eliminamos las filas donde los municipios no esten asignados
postal_code_canada_new = postal_code_canada[postal_code_canada['Borough']!= 'Not assigned']
postal_code_canada_new= postal_code_canada_new.reset_index(drop=True)
print(postal_code_canada_new.head(20))
#Verificamos si aún existen municipios que no esten asignados, debe ser negativo
print(postal_code_canada_new[postal_code_canada_new['Neighborhood'] == 'Not assigned'])
```

```python
# Tamaño del dataframe.
print('El dataframe tiene {} filas'.format(postal_code_canada_new.shape[0]), 'y {} columnas'.format(postal_code_canada_new.shape[1]))
```

```python
import folium
from geopy.geocoders import Nominatim #librería para buscar las coordenadas
```

```python
#Buscamos las coordenadas geográficas de Toronto
address = 'Toronto, Ontario'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('Las coordenadas geográficas de Toronto son {}, {}.'.format(latitude, longitude))
```

```python
# crear un mapa de Toronto usando los valores de latitud y longitud
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# añadir los marcadores al mapa
for lat, lng, label in zip(df_postal_code_canada['Latitude'], df_postal_code_canada['Longitude'], df_postal_code_canada['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='black',
        fill=True,
        fill_color='blue',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
